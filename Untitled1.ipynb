```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0, 10.0)

#Reading data from .csv format file
data=pd.read_csv("C:/Users/Kshitij/Desktop/headbrain.csv")
print(data.shape)
data.head()

```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
#find mean of x and y
mean_x=np.mean(X)
mean_y=np.mean(Y)

#total no. of values
n=len(X)

#calculate b0 and b1
num=0
den=0

for i in range(n):
    num +=(X[i]-mean_x)*(Y[i]-mean_y)
    den +=(X[i]-mean_x)**2
    
b1=num/den
b0=mean_y-(b1*mean_x)

#print coefficients
print(b1, b0)
```

```python
#plotting values and regression line
max_x=np.max(X)+100
min_x=np.min(X)-100

#Calculating line values x and y
x=np.linspace(min_x, max_x, 1000)
y=b0+b1*x

#plotting line
plt.plot(x, y, color='#57c970', label='Regresion line')

#plotting point
plt.scatter(X, Y, color='#ef5423', label='point')

plt.xlabel('Head Size')
plt.ylabel('Brain weight')

plt.legend()
plt.show()
```

```python
#Using R^2 method to estimate error or accuracy of model
#ss_t total sum of square  

ss_t=0
ss_r=0
for i in range(n):
    yp=b0+b1*X[i]
    ss_t +=(Y[i]-mean_y)**2
    ss_r +=(Y[i]-yp)**2
    
r2=1-(ss_r/ss_t)
print(r2)
```
